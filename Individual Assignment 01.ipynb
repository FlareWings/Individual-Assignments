#### T1. Read the dataset to python environment.

```python
# Importing Pandas Library as 'pd'
import pandas as pd
# Copying the Data from a Microsoft Excel 97-2003 Worksheet to Dataframe
Iris_Data = pd.read_excel('iris.xls')
```

#### T2. Display the columns in the dataset.

```python
# Displaying the preview of column by including 5 random rows
Iris_Data.sample(5)
```

#### T3. Calculate the mean of each column of the dataset.

```python
# Getting the mean of the required columns by passing the string
Mean = Iris_Data[['SL','SW','PL','PW']].mean()
print('The Data given below are the mean value of each column in the Dataframe.')
Mean
```

#### T4. Check for the null values present in the dataset.

```python
# Getting the Information Like Memory usage, Non-null counts etc... (null_counts depreciated)
Iris_Data.info()
# In-Depth view of null values in the Dataframe (Dedicated scrollable view)
Iris_Data.isnull()
```

#### T5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
# 1.Relational plot
# Importing Seaborn Library as 'sns'
import seaborn as sns
sns.relplot(x="PL", y="SL", kind="line", data=Iris_Data, hue="Classification")
sns.relplot(x="PW", y="SW", kind="line", data=Iris_Data, hue="Classification")
```

```python
# 2.Pairplot
ID = sns.PairGrid(Iris_Data, hue="Classification")
ID.map_upper(sns.scatterplot)
ID.map_lower(sns.kdeplot)
ID.map_diag(sns.kdeplot, lw=0.75, legend=False)
```

```python
# 3.Box plot
sns.boxplot(x="Classification", y="PL", hue="Classification", data=Iris_Data)
```

```python
sns.boxplot(x="Classification", y="SL", hue="Classification", data=Iris_Data)
```

```python
# 4.Scatter plot
# Importing Pyplot from Matplot Library as 'plt'
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(Iris_Data['PL'],Iris_Data['SL'],s=10,marker='*',c='#53d769')
plt.title('Scatterplot of Petal length Vs Sepal length', fontsize=15)
plt.xlabel('Petal length', fontsize=10)
plt.ylabel('Sepal length', fontsize=10)
```

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(Iris_Data['PW'],Iris_Data['SW'],s=10,marker='*',c='#4285f4')
plt.title('Scatterplot of Petal width Vs Sepal width', fontsize=15)
plt.xlabel('Petal width', fontsize=10)
plt.ylabel('Sepal width', fontsize=10)
```
