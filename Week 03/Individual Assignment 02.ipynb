## Task 1. Read the dataset to the python environment.

```python
# Pandas Library as 'pd'
import pandas as pd
# Writing the Dataset from the csv file to Pandas Dataframe named 'cd'
cd = pd.read_csv('cars_data.csv')
```

## Task 2. Check for the null values present in the dataset.

```python
# Getting the null value status of the Dataframe using 'dataframe.isnull()'
cd.isnull() # 💡 We can use cd.isna() also! 
```

> ### *From this information pane it is clear that the data of 10000 car entries with 20 categories are non-null in nature.*

```python
# Getting the null value count from the Dataframe using 'dataframe.isnull().sum()'
cd.isnull().sum() # 💡 Alternative method 'cd.isna().sum()'
```

> ### *There is no null value present in any of the columns/categories in this dataframe.*

### Task 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (8,8))
sns.countplot(x='Buyer Gender',hue='Buyer Gender',data=cd)
plt.xlabel("Gender", fontsize='12.5')
plt.ylabel("Count", fontsize='12.5')
plt.title("Count of both Male and Female buyers", fontsize='15')
plt.legend(loc=(1.04,0),title="Buyer Gender",prop={'size': 15})
plt.grid()
plt.show()
```

> ### *There is only negotiable difference in buyer count based on gender.*

### Task 4. Find the top 5 cars based on their sales price.

```python
# Using cd.nlargest to get Top 5 entries in the Database and assigning to 'cdl'
cdl = cd.nlargest(5, ['Sale Price'])
cdl
```

```python
# Locating the 3 Categories using cdl.loc
cdl.loc[: ,['Make','Model','Sale Price']]
```

> ### Insights
> - Most of the top-tier buyers prefer used cars.
> - Only 2 of them are in Luxury car segment, rest of the 3 are in Sports car segment. Which means capable buyers give equal importance to sports car segment in Top-tier along with Luxury cars.

> 

### Task 5. Find the least 5 cars based on their Resell price.

```python
# Using cd.nsmallest to get bottom 5 entries in the Resell Price category and assigning to 'cds'
cds = cd.nsmallest(5, ['Resell Price'])
cds
```

```python
# Locating the 3 Categories using cds.loc
cds.loc[: ,['Make','Model','Sale Price']]
```

> ### *Performance car division vitnessed a much lower band of reselling price.*
